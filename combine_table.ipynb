```python
def table_to_dataframe(latex_table):
    # Replace LaTeX-specific formatting with CSV-friendly formatting
    latex_table = latex_table.replace('&', ',').replace('\\hline', '').replace('\\\\', '\n').replace('$', '').replace('\\text{', '').replace('}', '').replace('{', '').strip()
    
    # Read the modified string into a DataFrame
    data = StringIO(latex_table)
    df = pd.read_csv(data)
    
    df['function value'] = df.index.get_level_values(1)
    df.index = df.index.get_level_values(0)
    df.columns = ['gradient norm', 'iter', 'runtime', '1', '2', '3', '4', '6', 'function value']
    df = df[['function value', 'gradient norm', 'iter', 'runtime']]
    return df 


df1 = table_to_dataframe(table_p1)
df2 = table_to_dataframe(table_p2)
df3 = table_to_dataframe(table_p3)
df4 = table_to_dataframe(table_p4)
df5 = table_to_dataframe(table_p5)
df6 = table_to_dataframe(table_p6)
df7 = table_to_dataframe(table_p7)
df8 = table_to_dataframe(table_p8)
df9 = table_to_dataframe(table_p9)
df10 = table_to_dataframe(table_p10)
df11 = table_to_dataframe(table_p11)
df12 = table_to_dataframe(table_p12)

df2
```
